```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('/content/new.csv')

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(df['Step'], df['Training'], marker='o', linestyle='-', color='b')
plt.title('Training Loss vs Step(accuracy = 0.84268)')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
```
